```python
import pandas as pd
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.linear_model import LogisticRegression
import joblib
```

```python
# abrindo os metadados
df = pd.read_csv("./../../../meta_dados/input_data/experiment_2/unsupervised_metric_sample_2.csv")
X = df.drop(columns=["target"]).to_numpy()
y = df["target"].to_numpy()
```

```python
# criando kfold
kf = KFold(n_splits=60, shuffle=True, random_state=42)

# 2. Grid de parmetros mais prtico
param_grid = [
    {   'solver': ['lbfgs', 'sag', 'newton-cholesky', 'newton-cg'],
        'penalty': [ 'l2', None],
        'C': [0.01, 0.1, 1, 3],
        'fit_intercept': [True, False]
    },
    {
        'solver': ['saga'],
        'penalty': ['l1', 'l2', 'elasticnet', None],
        'C': [0.01, 0.1, 1, 3],
        'fit_intercept': [True, False],
    }
]

lr = LogisticRegression(random_state=42)


# Configurar o GridSearchCV
grid_search = GridSearchCV(
    estimator=lr,
    param_grid=param_grid,
    cv=kf, 
    scoring='accuracy', 
    n_jobs=-1,  
    verbose=1
)

# Executar o GridSearchCV
grid_search.fit(X, y)
```

```python
# Melhores parmetros encontrados
print("Melhores parmetros:", grid_search.best_params_)

# Melhor score
print("Melhor acurcia:", grid_search.best_score_)
```

```python
# Salvar modelo
joblib.dump(grid_search.best_estimator_, './saved_models/lr.joblib')
```
